## Create Data from Existing data 

```python
car_sales
```

```python
seats_column = pd.Series([5,5,4,5,8])
car_sales["Seats"] = seats_column
car_sales["Seats"].fillna(6, inplace = True)
car_sales
```

```python
# Column from Python list; it has to be of same length as your dataframe...
fuel_economy = [7.5, 9.2, 6.0, 4.5, 8.7, 2.1, 8.0, 9.5, 11.5, 7.6]
car_sales["Fuel per 100KM"] = fuel_economy
car_sales
```

```python
car_sales["Total Fuel used"] = car_sales["Odometer (KM)"]/100 * car_sales["Fuel per 100KM"]
car_sales
```
